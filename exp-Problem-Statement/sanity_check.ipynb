```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
from IPython.display import display
import itertools, time, os, pickle, json, tempfile, sys
srcpath = os.path.abspath(os.path.join(os.path.abspath(''),  '..',  'src'))
sys.path.append(srcpath)

from exp_utils import *
#from data_gen_full import *




plt.rcParams.update({
     "text.usetex" : "true",
    "font.family": "serif",
    "font.serif": ["Palatino"],
})


plt.rc('figure', dpi=150)

acc = lambda x,y: 1 - np.mean(0.5*np.abs(x.flatten()-y.flatten()))

with open('data_full_new.pkl', 'rb') as f:
    data = pickle.load(f)
    data = pd.DataFrame.from_dict(data, orient = 'index')
    data.index = pd.MultiIndex.from_tuples(data.index,  names=['kernel,dataset', 'm_train', 'm_test',
                                                               'C', 'N' ,'mode'])
data = data.xs( (1, 60, 60,  10, 'test_train'), level = ['C', 'm_train', 'm_test','N','mode'] )
data = data[["K_train", "K_test_train", "y_train", "y_test"]]
```

```python
## Usage
##Copying the kernels and labels into a nested dict fo sanity checking,
##The structure of the dic is {key: {K_train:_ , K_test_train:_, y_train:_ , y_test_train:_}
##So  data_dict[key]["K_train"] will retrun the K_train for that key.

data_dict = data.to_dict(orient = "index")

for k in data_dict:
    C = 1
    v= data_dict[k]
    K_train, K_test_train, y_train, y_test = v["K_train"], v["K_test_train"], v["y_train"], v["y_test"]
    clf =  SVC(kernel='precomputed', C = C)
    clf.fit(K_train, y_train)

    print(k)
    print(f"Train_Acc:{acc(y_train, np.sign(clf.decision_function(K_train)))}")
    print(f"Test_Acc:{acc(y_test, np.sign(clf.decision_function(K_test_train)))}")
    print("____________________________________________________________________")
    print("")
```
